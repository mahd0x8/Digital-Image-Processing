Logrithmic Transformation

````python
def Log(Obj,C=1):
  Obj1 = Obj.copy()
  for i in range(len(Obj1)):  
    for j in range(len(Obj1[0])):
      r = Obj1[i][j]
      Obj1[i][j][0] = round(C*np.log(r[0]+1))
      Obj1[i][j][1] = round(C*np.log(r[1]+1))
      Obj1[i][j][2] = round(C*np.log(r[2]+1))
  return Obj1
````

````python
plt.imshow(Log(Img,20), interpolation='nearest')
plt.show()
````

Power Transformation or Gamma Fixing

````python
def Power(Obj,C=1,Y=1):
  Obj1 = Obj.copy()
  for i in range(len(Obj1)):  
    for j in range(len(Obj1[0])):
      r = Obj1[i][j]
      Obj1[i][j][0] = round(C*(r[0]**Y))
      Obj1[i][j][1] = round(C*(r[1]**Y))
      Obj1[i][j][2] = round(C*(r[2]**Y))
  return Obj1
````

````python
plt.imshow(Power(Img,C = 5,Y=0.7), interpolation='nearest')
plt.show()
````
